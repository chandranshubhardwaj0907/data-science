```python
import pandas as pd
import numpy as np

# Number of rows
n = 10000

# Generate data
age = np.random.randint(18, 70, n)
estimated_price = np.random.randint(20000, 70000, n)
purchased = np.random.randint(0, 2, n)  # 0 or 1

# Create DataFrame
df = pd.DataFrame({
    'Age': age,
    'EstimatedPrice': estimated_price,
    'Purchased': purchased
})

# Save to CSV
df.to_csv('social_network_ads_big.csv', index=False)

# Show first 5 rows
print(df.head())
```
