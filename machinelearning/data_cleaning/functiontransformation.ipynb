```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('loan.csv')
```

```python
data
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
data['CoapplicantIncome_tt'] = FunctionTransformer(func=np.log1p).fit_transform(data[['CoapplicantIncome']])
```

```python

plt.figure(figsize=(10,5))  # 'figure' not 'Figure'

# First subplot
plt.subplot(1,2,1)
sns.histplot(data['CoapplicantIncome'], kde=True)  # use histplot instead of displot for subplots
plt.title('Before Scaling')

# Second subplot
plt.subplot(1,2,2)
sns.histplot(data['CoapplicantIncome_tt'], kde=True)
plt.title('After Scaling')

plt.tight_layout()
plt.show()
```

```python
data['CoapplicantIncome_xx'] = FunctionTransformer(func=lambda x : x**2).fit_transform(data[['CoapplicantIncome']])
```

```python

plt.figure(figsize=(10,5))  # 'figure' not 'Figure'

# First subplot
plt.subplot(1,2,1)
sns.histplot(data['CoapplicantIncome'], kde=True)  # use histplot instead of displot for subplots
plt.title('Before Scaling')

# Second subplot
plt.subplot(1,2,2)
sns.histplot(data['CoapplicantIncome_xx'], kde=True)
plt.title('After Scaling')

plt.tight_layout()
plt.show()
```
