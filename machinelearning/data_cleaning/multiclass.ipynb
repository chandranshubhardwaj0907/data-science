```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
import pandas as pd
import numpy as np

# Number of students
n = 1000

# Generate CGPA between 5.0 and 10.0
cgpa = np.round(np.random.uniform(5.0, 10.0, n), 2)

# Generate Score between 50 and 100
score = np.round(np.random.uniform(50, 100, n), 2)

# Generate placement status (1 = placed, 0 = not placed)
# Higher CGPA and higher Score increases chance of placement
placed = ((cgpa * 10 + score + np.random.uniform(-20, 20, n)) > 120).astype(int)

# Create DataFrame
df = pd.DataFrame({
    'CGPA': cgpa,
    'Score': score,
    'Placed': placed
})

# Save to CSV
df.to_csv('cgpa_score_placed_dataset.csv', index=False)

# Show first 10 rows
print(df.head(10))
```

```python
sns.scatterplot(x ='CGPA' , y = 'Placed' , data = df , hue = 'Placed')
```

```python
x = df.iloc[:,:-1]
y = df['Placed']
```

```python
x_train , x_test  , y_train , y_test = train_test_split(x,y , test_size=0.2  , random_state=42)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train , y_train)
```

```python
lr.score(x_test , y_test)*100
```

```python
lr.predict([[8.53 , 88.61]])
```
