```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
seca = np.array([75,65,73,68,72,76])
secb = np.array([90,47,43,96,93,51])
no = np.array([1,2,3,4,5,6])
```

```python
mean = np.mean(secb)
```

```python
mean.astype(int)
```

```python
plt.Figure(figsize=(3,8))
plt.scatter(seca , no , label = "seca")
plt.scatter(secb , no , label = "secb")
plt.axvline(mean , color = "red" , label = "mean")
plt.legend()
plt.show()
```

```python
mada = np.sum(np.abs(seca-mean))/len(seca)
```

```python
madb = np.sum(np.abs(secb-mean))/len(secb)
```
