```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# using minmax scaler to scale the data between 0 and 1
```

```python
data = pd.read_csv('loan.csv')  
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
```

```python
mm.fit(data[['CoapplicantIncome']])
```

```python
mm.transform(data[['CoapplicantIncome']])   
```

```python
data['CoapplicantIncome_mm'] = pd.DataFrame(mm.transform(data[['CoapplicantIncome']]) , columns=['x']) 
```

```python
data
```

```python
plt.figure(figsize=(20,6))

plt.subplot(1,2,1)
plt.title('Before Scaling')
sns.histplot(data['CoapplicantIncome'], kde=True)

plt.subplot(1,2,2)
plt.title('After Scaling')
sns.histplot(data['CoapplicantIncome_mm'], kde=True, color="orange")

plt.tight_layout()
plt.show()
```
